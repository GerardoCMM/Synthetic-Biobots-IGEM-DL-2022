```python
import pandas as pd
```

```python
up40 = pd.read_csv("Clusters_dif40.csv")
```

```python
AHRD = up40["AHRD.Descriptions"]
TAIR = up40["TAIR.Gene.s..Descriptions"]
domains = up40["Pfam.Domains"]
funcs = up40["GO.Molecular.Functions"]
```

```python
AHRD2 = []
for item in AHRD:
    AHRD2.append(item.split("[")[0])
    
TAIR2 = []
for item in TAIR:
    TAIR2.append(item.split("|")[0])
    
domains2 = []
for item in domains:
    domains2.append(item.split("(")[0])
    
funcs2 = []
for func in funcs:
    funcs2.append(func.split("(")[0])
```

```python
up40["AHRD.Descriptions"] = AHRD2
up40["TAIR.Gene.s..Descriptions"] = TAIR2
up40["Pfam.Domains"] = domains2
up40["GO.Molecular.Functions"] = funcs2
```
